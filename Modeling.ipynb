```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, roc_auc_score, roc_curve, auc, accuracy_score, precision_score, recall_score 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost.sklearn import XGBClassifier  
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from scipy import stats

import xgboost as xgb
import imblearn.over_sampling
import matplotlib.pyplot as plt
import matplotlib.style as style 
%matplotlib inline
import seaborn as sns
style.use('fivethirtyeight')
```

### Split train and test data so the same users are not in both sets

```python
df_X = pd.read_csv('instacart_df_X_features3.csv') 
   
np.random.seed(42)
total_users = df_X['user_id'].unique() 
test_users = np.random.choice(total_users, size=int(total_users.shape[0] * .20))

df_X_tr, df_X_te = df_X[~df_X['user_id'].isin(test_users)], df_X[df_X['user_id'].isin(test_users)]

y_tr, y_te = df_X_tr['in_cart'], df_X_te['in_cart']
X_tr, X_te = df_X_tr.drop(['product_id','user_id','latest_cart','in_cart'],axis=1), \
             df_X_te.drop(['product_id','user_id','latest_cart','in_cart'],axis=1), \
```

### Model for simple logistic regression and balanced/weighted Regression

```python
lr = LogisticRegression(solver='liblinear') 
lr_balanced = LogisticRegression(class_weight='balanced', solver='liblinear')
lr_5x = LogisticRegression(class_weight={1 : 5, 0 : 1}, solver='liblinear')
```

```python
lr.fit(X_tr, y_tr)
print('Normal Logistic Regression Test F1: %.3f, Test AUC: %.3f' % \
      (f1_score(lr.predict(X_te), y_te), roc_auc_score(y_te, lr.predict_proba(X_te)[:,1])))

lr_balanced.fit(X_tr, y_tr)
print('Balanced class weights Logistic Regression Test F1: %.3f, Test AUC: %.3f' % \
      (f1_score(lr_balanced.predict(X_te), y_te), roc_auc_score(y_te, lr_balanced.predict_proba(X_te)[:,1])))

lr_5x.fit(X_tr, y_tr)
print('5:1 class weights Logistic Regression Test F1: %.3f, Test AUC: %.3f' % \
      (f1_score(lr_5x.predict(X_te), y_te), roc_auc_score(y_te, lr_5x.predict_proba(X_te)[:,1])))
```

```python
X_val, y_val = X_te, y_te # explicitly calling this validation since we're using it for selection

thresh_ps = np.linspace(.1,.5,1000)
model_val_probs = lr.predict_proba(X_val)[:,1] # positive class probs, same basic logistic model we fit in section 2 

f1_scores = []
for p in thresh_ps:
    model_val_labels = model_val_probs >= p
    f1_scores.append(f1_score(model_val_labels, y_val))
    
plt.plot(thresh_ps, f1_scores)
plt.title('Logistic RegressionF1 Score vs. Positive Class Decision Probability Threshold')
plt.xlabel('P threshold')
plt.ylabel('F1 score')
plt.tight_layout()
plt.savefig('initial_lr.png')

best_f1_score = np.max(f1_scores) 
best_thresh_p = thresh_ps[np.argmax(f1_scores)]

print('Logistic Regression Model best F1 score %.3f at prob decision threshold >= %.3f' 
      % (best_f1_score, best_thresh_p))
```

```python
def get_F1(predictions_proba, y_test, thresh):
    y_pred = (predictions_proba > thresh).astype(int)
    
    return f1_score(y_pred, y_test)
```

```python
# Start
#Minimum loss reduction to further partition leaf node. Larger gamma is more conservative
# gamma=0: F1:.4002 <--
# gamma=100: F1:.39
 
#L2 Regularization, higher is more conservative
#lambda=1: F1:.4002
#lambda=1000: F1:.4023<--

#L1 Regularization, higher is more conservative
#alpha=0: F1: .4023<--
#alpha=100: F1: .4002

# Max depth of tree
#max_depth=3: F1: .4023<--
#max_depth=6: F1: .4002

#step size shrinkate
#learning_rate=.1: F1: .4023269<---
#learning_rate=.6: F1: .4002

#Min sum of instance weight needed in a child, larger is more conservative
#min_child_weight=3: F1: .4023
#min_child_weight=100: F1: .40328<---

#Subsample ratio of columns for each split, in each level
#colsample_bytree=.8: F1: .40328
#colsample_by_tree=.1: F1: .4039

# Best F1: .4039

gbm = xgb.XGBClassifier( 
                        n_estimators=100,
                        gamma=0,
                        reg_lambda=1000,
                        alpha=0,
                        max_depth=3,
                        objective='binary:logistic', #new objective
                        learning_rate=.1, 
                        min_child_weight=100,
                        colsample_bytree=.1,
                        tree_method='exact'
                       )
eval_set = [(X_te, y_te)]

gbm.fit(X_tr, y_tr, 
        early_stopping_rounds=15, 
        eval_metric=["logloss"], 
        eval_set=eval_set)

def get_best_f1(model, feat):
    f1_scores = []
    thresh = np.linspace(0.15,0.3,50)
    xgb_test_prob = model.predict_proba(X_te)[:,1]
    f1 = [get_F1(xgb_test_prob, y_te, t) for t in thresh]
    f1_scores.append(f1)
    best_thresh = thresh[np.argmax(f1_scores)]
    print(f'XG Boost Model best F1 score {max(f1)}, at decision threshold = {best_thresh})')
get_best_f1(gbm, X_te.columns)
```
